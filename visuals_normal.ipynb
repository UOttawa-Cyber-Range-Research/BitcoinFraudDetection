```python
def draw_plot(file_num, dataset, z):
    '''
    Plot files for each of the datasets
    '''
    # Load the file and stuff as per the dataset
    edge_file = f'./sampling_comparison/{dataset}_edges_{file_num}.parquet'
    label_file = f'./sampling_comparison/{dataset}_labels_{file_num}.parquet'
    
    # Load the files
    df_e = pd.read_parquet(edge_file)
    df_l = pd.read_parquet(label_file)
    print(df_l[df_l["label"] == 1])

    # Define the node colors
    color_map = []
    for _, row in df_l.iterrows():
        if int(row["label"]) == 1:
            color_map.append("red")
        else:
            color_map.append("blue")

    # Plot the graph
    
    G = nx.Graph()
    for index, row in df_e.iterrows():
        G.add_edge(row['from'], row['to'])

    print(f"Total edges : {G.number_of_edges()}")
    plt.subplot(1, 2, z)
    nx.draw(G, with_labels=False, node_size=60, node_color=color_map)
```

```python
p=2
plt.figure(figsize=(10, 6))
draw_plot(p, "bfs", 1)
draw_plot(p, "bf", 2)
plt.show()
```

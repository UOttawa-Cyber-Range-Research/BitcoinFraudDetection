```python
def draw_plot(file_num, dataset):
    '''
    Plot files for each of the datasets
    '''
    # Load the file and stuff as per the dataset
    edge_file = f"./data/dataset{dataset}/cache/edges/edges_{file_num}.parquet"
    label_file = f"./data/dataset{dataset}/labels/labels_{file_num}.parquet"
    
    # Load the files
    df_e = pd.read_parquet(edge_file)
    df_l = pd.read_parquet(label_file)

    # Define the node colors
    color_map = []
    for _, row in df_l.iterrows():
        if int(row["label"]) == 1:
            color_map.append("yellow")
        else:
            color_map.append("blue")

    # Plot the graph
    plt.figure(figsize=(6, 6))
    G = nx.Graph()
    for index, row in df_e.iterrows():
        G.add_edge(row['from'], row['to'])
    nx.draw(G, with_labels=False, node_size=60, node_color=color_map)
    plt.show()
```

```python
draw_plot(2, "RFS")
draw_plot(8, "RFS")
```

```python
draw_plot(2, "BFS")
draw_plot(8, "BFS")
```

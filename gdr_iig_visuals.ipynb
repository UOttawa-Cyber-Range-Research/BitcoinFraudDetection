```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataframes
test_df_bfron = pd.read_csv("../data/df/rggcn_datasetBFRON_GN_false_normal_test.csv")
test_df_mhrw = pd.read_csv("../data/df/rggcn_datasetMHRW_GN_false_normal_test.csv")
test_df_bfs = pd.read_csv("../data/df/rggcn_datasetBFS_GN_false_normal_test.csv")
test_df_fs = pd.read_csv("../data/df/rggcn_datasetFS_GN_false_normal_test.csv")
test_df_rfs = pd.read_csv("../data/df/rggcn_datasetRFS_GN_false_normal_test.csv")
```

```python
def plot_diagram(metric_value, sharey=True, show_rwpe=False):
    # Define the figure and axes
    fig, axes = plt.subplots(nrows=1, ncols=5, sharey=sharey, sharex=True, figsize=(24, 4))

    # Set the axis
    labels = [i for i in range(11)]
    plt.xticks(labels)
    
    # All the data
    sns.lineplot(data=test_df_bfron, x="num_layers", y=metric_value, marker='o', ax=axes[0])
    axes[0].set_title(f'BFRON vs {metric_value.upper()}')
    axes[0].set_xlabel('Num Layers')
    axes[0].set_ylabel(metric_value.upper())
    
    # Data no random walk positional embedding
    sns.lineplot(data=test_df_mhrw, x="num_layers", y=metric_value, marker='o', ax=axes[1])
    axes[1].set_title(f'MHRW vs {metric_value.upper()}')
    axes[1].set_xlabel('Num Layers')
    axes[1].set_ylabel(metric_value.upper())
    
    # Data random walk positional embedding
    sns.lineplot(data=test_df_bfs, x="num_layers", y=metric_value, marker='o', ax=axes[2])
    axes[2].set_title(f'BFS vs {metric_value.upper()}')
    axes[2].set_xlabel('Num Layers')
    axes[2].set_ylabel(metric_value.upper())
    
    # Data random walk positional embedding
    sns.lineplot(data=test_df_fs, x="num_layers", y=metric_value, marker='o', ax=axes[3])
    axes[3].set_title(f'FS vs {metric_value.upper()}')
    axes[3].set_xlabel('Num Layers')
    axes[3].set_ylabel(metric_value.upper())

    # Data random walk positional embedding
    sns.lineplot(data=test_df_rfs, x="num_layers", y=metric_value, marker='o', ax=axes[4])
    axes[4].set_title(f'RFS vs {metric_value.upper()}')
    axes[4].set_xlabel('Dataset')
    axes[4].set_ylabel(metric_value.upper())
    
    # Show the plot
    plt.show()
```

```python
plot_diagram("gdr", sharey=False)
```

```python
plot_diagram("iig", sharey=False)
```
